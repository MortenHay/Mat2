```python
n = Symbol('n',integer=True,positive=True)
x = Symbol('x')
f = 2**n/(4*n**2+2*n+1)*x**n
f2 = abs(f.subs(n,n+1)/f).simplify()
solve(GreaterThan(1,f2.limit(n,oo)))
```

```python
t = Symbol('t',real=True)
c = Function('c')(n)
y = Sum(c*t**n,(n,0,oo))
deq = Eq(y.diff(t,2)+t*y.diff(t,1)+2*y,t**2*exp(t))
deq
```
