Med stning 4.38 fs at rkkerne er konvergente hvis og kun hvis hvert led er absolut mindre end det foregende. Frst for rkken S. For denne rkke er $b_n$ svarende til definition 4.37
$$ b_n = \frac{\sqrt{n}}{n+3}$$
Denne talflge undersges med sammenligningskriteriet (stning 4.20) som
$$\frac{\sqrt{n}}{n+3} < \frac{\sqrt{n}}{n} = \frac{1}{\sqrt{n}}$$
$\frac{1}{\sqrt{n}}$ aftager monotont, s rkken er konvergent. Hvorvidt den er absolut konvergent undersges med kvivalenskriteriet med $a_n = \frac{1}{\sqrt{n}}$
$$ \frac{\frac{1}{\sqrt{n}}}{\frac{\sqrt{n}}{n+3}} = \frac{n+3}{n} \rightarrow 1~\mathrm{for~n} \rightarrow \infty$$
Rkkerne er alts kvivalente. $\sum_{n=1}^{\infty}\frac{1}{\sqrt{n}}$ er divergent, s det er $\sum_{n=1}^{\infty}\frac{\sqrt{n}}{n+3}$ ogs. Alts er S betinget konvergent.  
T kan undersges ved at omskrive den til en kvotientrkke
$$ \sum_{n=1}^{\infty}(-1)^n\frac{3^{n+1}}{2^{2n}} = \sum_{n=1}^{\infty}(-1)^n3\frac{3^{n}}{4^{n}} = 3\sum_{n=1}^{\infty}(-1)^n\left(\frac{3}{4}\right)^n$$
Med stning 5.2 er rkken $\sum_{n=1}^{\infty}\left(\frac{3}{4}\right)^n$ konvergent. Alts er T absolut konvergent

Ved indsttelse i systemet med $\mathbf{v}= \begin{pmatrix}v_1\\v_2\end{pmatrix}$ fs
$$(-te^{-t}+e^{-t})\begin{pmatrix}v_1\\v_2\end{pmatrix} = \begin{pmatrix}-1&0\\-1&-2\end{pmatrix}te^{-t}\begin{pmatrix}v_1\\v_2\end{pmatrix}+e^{-t}\begin{pmatrix}1\\-1\end{pmatrix}$$  
$$\Rightarrow \begin{pmatrix}0\\v_1+v_2\end{pmatrix}te^{-t}+\begin{pmatrix}v_1\\v_2\end{pmatrix}e^{-t} = e^{-t}\begin{pmatrix}1\\-1\end{pmatrix}$$  
Hvis dette skal overholdes for alle $t\in\mathbb{R}$ skal koefficienter for hver orden af t vre ens p begge sider:
$$\begin{pmatrix}0\\v_1+v_2\end{pmatrix} = \begin{pmatrix}0\\0\end{pmatrix}$$
$$\begin{pmatrix}v_1\\v_2\end{pmatrix} = \begin{pmatrix}1\\-1\end{pmatrix}$$
Hvis det skal vre en lsning skal $v_1= 1,v_2=-1$ Dette indsttes i $v_1 + v_2 = 1 + (-1) = 0$ hvilket overholder ligningerne. Alts er der en partikulr lsning
$$x_p(t) = te^{-t}\begin{pmatrix}1\\-1\end{pmatrix}$$

Den fuldstndige lsning til systemet kan nu opskrives med stning 1.20 som  
$$x(t)=x_{hom}(t)+x_p(t)$$  
$$x(t)=c_1e^{-t}\begin{pmatrix}-1\\1\end{pmatrix}+c_2e^{-2t}\begin{pmatrix}0\\1\end{pmatrix}+te^{-t}\begin{pmatrix}1\\-1\end{pmatrix},~~c_1,c_2\in\mathbb{R}$$

$$a_n=\frac{2}{\pi}\int_0^\pi \cos\left(\frac{3}{2}t\right) =  \frac{2}{\pi}\frac{3}{2}\frac{\sin\left(\pi\frac{3}{2}\right)\cos\left(n\pi\right)}{\frac{9}{4}-n^2}=\frac{12}{\pi}\frac{-1\cos\left(n\pi\right)}{9-4n^2}$$

Fourierrkken opskrives med korollar 6.7 $(i)$  
$$f \sim \frac{1}{2}a_0+\sum_{n=1}^{\infty}a_n\cos(nx)=-\frac{2}{3\pi}+\sum_{n=1}^{\infty}\frac{12(-1)^n}{\pi(4n^2-9)}\cos(nx)$$


Hvorvidt rkken er kontinuert skal undersges. Da cosinus i sig selv er en kontinuert funktion, vil $f$ kun kunne vre kontinuert hvor den gentages. Da funktionen er lige, m $f(-t) = f(t) \Leftarrow f(0) = f(-0)$ s funktionen er kontinuert i $t=0$. I $t=\pi$ hvor funktionen gentages m det ligeledes glde, at $f(\pi)=f(-\pi)=f(-\pi + 2\pi)$. Funktionen er alts ogs kontinert i endepunkterne. $f$ er alts kontinuert i alle $t\in\mathbb{R}$. Med korollar 6.17 $(ii)$ er den derved ogs uniformt konvergent. 

Der kan findes en majorantrkke for $f$ med definition 5.31. Her sttes $f_n(x) = \frac{12(-1)^n}{\pi(4n^2-9)}\cos(nx)$. Da bliver  
$$\left|f_n(x)\right| = \left|\frac{12(-1)^n}{\pi(4n^2-9)}\cos(nx)\right|= \frac{12}{\pi}\left|\frac{\cos(nx)}{4n^2-9}\right| \leq\frac{12}{\pi}\frac{1}{4n^2-9},~\mathrm{for}~n\geq 2$$
Alts er der en Majorantrkke for $f$ givet ved  
$$\frac{12}{\pi}\sum_{n=2}^{\infty}\frac{1}{4n^2-9}$$  
Denne er kvivalent med $\sum_{n=2}^{\infty}\frac{1}{n^2}$, da
$$\frac{\frac{1}{n^2}}{\frac{1}{4n^2-9}} = \frac{4n^2-9}{n^2}\rightarrow 4=C~\mathrm{for}~n\rightarrow \infty$$  
Da $\sum_{n=2}^{\infty}\frac{1}{n^2}$ er konvergent, m majorantrkken ogs vre konvergent.
