Ud fra karakterligningen kan der aflses rdderne $\lambda = -1,\pm i$ alle med algebrarisk multiplicitet 1. Derved kan den fuldstndige lsning opskrives med stning 1.15:  
$$y(t) = c_1e^{-t} + c_2\cos(t) + c_3\sin(t),~~c_1,c_2,c_3\in\mathbb{R}$$

Overfringsfunktionen kan bestemmes som lrebogens ligning (1.23) i eksempel 1.24. Dog kan nvneren erstattes med venstresiden fra den givne karakterligning for det homogene system.  
$$H(s)=\frac{1}{(s^2+1)(s+1)},~~ s\neq -1,\pm i$$

Det stationre svar kan findes ved at stte $u(t) = \mathrm{Re}(e^{2it})$. Da kan det stationre svar findes med stning 1.25 og 1.27 som
$$y(t) = \mathrm{Re}(H(2i)e^{2it})$$  
$$y(t) = \mathrm{Re}\left(\frac{1}{(-4+1)(2i+1)}\right)$$  
$$y(t) = \mathrm{Re}\left(\frac{1}{15}(-1+2i)(\cos(2t) + i\sin(2t))\right) $$  
$$y(t) = -\frac{1}{15}\cos(2t)-\frac{2}{15}\sin(2t)$$

Der skal gttes p en lsningsform. Da $ce^{-t}$ er en lsning til den homogene ligning, kan denne form ikke bruges. Derfor gttes p en lsning p formen $y(t) = cte^{-t}$ Dette indsttes i differentialligningen.

```python
#Symboler defineres
c = Symbol('c')
t = Symbol('t',real=True)

#Lsningsgttet defineres
y = c*t*exp(-t)

#Den inhomogene diffentialligning opskrives
deq = Eq(y.diff(t,3)+y.diff(t,2)+y.diff(t)+y,exp(-t))

#Den udskrives
deq
```

Udtrykket kan reduceres ved at sttes $e^{-t}$ ud for en parantes og koefficienterne kan reduceres ned til

```python
deq.simplify()
```

Af dette kan det let aflses, at $c=\frac{1}{2}$ og den partikulre lsning bliver s $y(t)=\frac{1}{2}te^{-t}$

Den fuldstndige lsning findes med stning 1.20 samt superpositionsprincippet stning 1.23. Lsningen til den homogene ligning blev fundet til  
$$y_{hom}(t) = c_1e^{-t} + c_2\cos(t) + c_3\sin(t),~~c_1,c_2,c_3\in\mathbb{R}$$  
En partikulr lsning er med superpositionsprincippet summen af den stationre lsning fra delopgave (iii) samt den partikulre lsning fra delopgave (iv)  
$$y_0 = \frac{1}{2}te^{-t} -\frac{1}{15}\cos(2t)-\frac{2}{15}\sin(2t)$$
Den fuldstndige relle lsning er s
$$y(t) = y_{hom}(t) + y_0(t)$$  
$$y(t) =  c_1e^{-t} + c_2\cos(t) + c_3\sin(t) + \frac{1}{2}te^{-t} -\frac{1}{15}\cos(2t)-\frac{2}{15}\sin(2t),~~c_1,c_2,c_3\in\mathbb{R}$$

Summen af fourerrkken kan findes i alle punkter med Fouriers stning (stning 6.16), for punkterne, der ikke ligger p en diskontinuitet er summen blot funktionesvrdien i punktet. Dette glder for punkterne
$$f_{kontinurt}\begin{cases}
f(-\pi) = 0\\
f(0) = -1\\
f\left(\frac{\pi}{2}\right) = 0\\
f(\pi) = 0
\end{cases}$$
I det diskontinurte punkt $x=-\frac{\pi}{2}$ er summen middelvrdien af grnsevrdierne i punktet fra hjre hhv. venstre. Alts er det
$$f(\frac{\pi}{2}) = \frac{0-2}{2} = -1$$  
Nu er summen af fourierrkken fundet i alle de nskede punkter

Funktionen er hverken lige eller ulige. Derfor skal fourierkoefficienter findes med definition 6.1
$$a_n = \frac{1}{\pi}\int_{-\pi}^{\pi}f(x)cos(nx)dx$$
$$b_n = \frac{1}{\pi}\int_{-\pi}^{\pi}f(x)sin(nx)dx$$

```python
n = Symbol('n',positive=True,integer=True)
a0 = (1/pi*integrate(f,(x,-pi,pi))).simplify()
an = (1/pi*integrate(f*cos(n*x),(x,-pi,pi))).simplify()
bn = (1/pi*integrate(f*sin(n*x),(x,-pi,pi))).simplify()
a0,an,bn
```

Da pvirkningen er et andengradspolynomie og venstresiden indeholder et ikkedifferentieret led, vil et passende gt vre et andengradspolinomie p formen $at^2+bt$ (Det kan let ses, at der aldrig kan vre et konstantled med den givne ligning)

```python
a,b = symbols('a b')
t = Symbol('t',real=True)
y = a*t**2+b*t
deq = Eq(t*diff(y,t,2)+y,t**2)
deq
```

Og derved  
$$\begin{matrix}

b = -2
\end{matrix}$$
Hvilket giver den partikulre lsning
$$y_0 = at^2-2t$$

Rkken differentieres med stning 5.17 og indsttes i differentialligningen
$$t\sum_{n=2}^{\infty}a_nn(n-1)t^{n-2}+\sum_{n=0}^{\infty}a_nt^n=0$$  
Der skal bestemmes en rekursionsformel for $a_n$  
$$\sum_{n=2}^{\infty}a_nn(n-1)t^{n-1}+\sum_{n=0}^{\infty}a_nt^n=0$$  
$$a_0 + \sum_{n=1}^{\infty}a_{n+1}n(n+1)t^{n}+\sum_{n=1}^{\infty}a_nt^n=0$$  
$$a_0 + \sum_{n=1}^{\infty}(a_{n+1}n(n+1)+a_n)t^n=0$$  
Med identitetsstningen for potensrkker (korollar 5.21) er dette opfyldt hvis og kun hvis  
$$a_0=0$$  
$$a_{n+1}n(n+1)+a_n = 0 $$
$$\Leftrightarrow \frac{a_{n+1}}{a_n} = -\frac{1}{n(n+1)} $$
